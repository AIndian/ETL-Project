```python
o_df = pd.read_csv("./Resources/documents_signed_by_barack_obama.csv")
t_df = pd.read_csv("./Resources/documents_signed_by_donald_trump.csv")
b_df = pd.read_csv("./Resources/documents_signed_by_joe_biden.csv")
all_df = pd.concat([o_df, t_df, b_df], axis=0, join='inner')

all_df["signing_date"]= pd.to_datetime(all_df['signing_date'], format = '%m/%d/%y')
all_df["signing_date"]= all_df["signing_date"].dt.strftime('%Y-%m-%d')
```
