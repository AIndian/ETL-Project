```python
import pandas as pd
import requests
from config import api_key
import pandas as pd
import json
from datetime import datetime as dt
```

```python
#base url for connecting to API
symbol = 'DIA'
url = f'https://www.alphavantage.co/query?function=TIME_SERIES_WEEKLY_ADJUSTED&symbol={symbol}&apikey={api_key}'
```

```python
response = requests.get(url)
print(response.url)
```

```python
data = response.json()
```

```python
print(json.dumps(data, indent=4, sort_keys=True))
```

```python
adjclose = []
close = []
date = []
vol = []
for time in data['Weekly Adjusted Time Series']:
    adjclose.append(data['Weekly Adjusted Time Series'][time]['5. adjusted close'])
    close.append(data['Weekly Adjusted Time Series'][time]['4. close'])
    vol.append(data['Weekly Adjusted Time Series'][time]['6. volume'])
    date.append(time)
```

```python
dia_df = pd.DataFrame({'date': date, 'adj. close': adjclose,
                      'close': close, 'vol': vol})
```

```python
dia_df['date']= pd.to_datetime(dia_df['date'], format = '%Y-%m-%d')
dia_df
```

```python
dia_df = dia_df.loc[dia_df['date']> '2009-01-05']
```
