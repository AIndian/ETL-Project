```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy_utils import database_exists, create_database
from config import code
```

```python
dia_file = '../finance/DIAdata.csv'
dia_df = pd.read_csv(dia_file)
dia_df = dia_df.drop(dia_df.columns[0], axis=1).set_index('date')
dia_df
```

```python
EO_file = '../all_df.csv'
EO_df = pd.read_csv(EO_file)
EO_df = EO_df.drop(EO_df.columns[[0]], axis=1)
EO_df=EO_df.set_index('signing_date')
EO_df
```

```python
EA_file = '../Main/appointments.csv'
EA_df = pd.read_csv(EA_file)
EA_df = EA_df.drop(EA_df.columns[[0, 1]], axis=1)
EA_df = EA_df.set_index('Service Start')
EA_df
```

```python
connection_string = f"postgres:2799@localhost:5432/presfin_db"
engine = create_engine(f'postgresql://{connection_string}', echo= True)
```

```python
if not database_exists(engine.url):
    create_database(engine.url)
else:
    # Connect the database if exists.
    engine.connect()
```

```python
dia_df.to_sql(name='DIA stock', con=engine, if_exists='replace', index=True)
EO_df.to_sql(name='orders', con=engine, if_exists='replace', index=True)
EA_df.to_sql(name='appointments', con=engine, if_exists='replace', index=True)
```
